# SQL - BigQuery

#### Data Exploration - **Chicago_crime** dataset from project **bigquery-public-data**. In BigQuery each dataset is contained in a corresponding project.

```python
#importing bigquery
from google.cloud import bigquery
```

```python
#create a 'Client' object, client object helps you to retriev information from Bigquery 
client=bigquery.Client()
```

```python
#construct a reference to the 'chicago_crime' dataset
dataset_ref=client.dataset('chicago_crime',project="bigquery-public-data")

#fetch the dataset
dataset=client.get_dataset(dataset_ref)
```

### Data Exploration

```python
#list all the tables in the chicago_crime dataset
tables=list(client.list_tables(dataset))

#print names of the table
for t in tables:
    print(t.table_id)
```

```python
#count table in the dataset
len(tables)
```

```python
#explore the table schema
table_ref=dataset_ref.table('crime')
table=client.get_table(table_ref)
table.schema
```

```python
#Preview the first 5 rows
client.list_rows(table,max_results=5).to_dataframe()
```

```python
#preview the first 5 rows in the 'date' column
client.list_rows(table,max_results=5,selected_fields=table.schema[2:3]).to_dataframe()
```

### FROM, WHERE

### COUNT()
return a count of things

```python
query4="""
    SELECT COUNT(unique_key)
    FROM `bigquery-public-data.chicago_crime.crime`
    """
rt4=client.query(query4).to_dataframe()
rt4
```

```python
# COUNT(1)- to count the rows in each group
query5="""
    SELECT COUNT(1) AS total
    FROM `bigquery-public-data.chicago_crime.crime`
    """
rt5=client.query(query5).to_dataframe()
rt5
```

```python
query6="""
    SELECT COUNT(case_number) 
    FROM `bigquery-public-data.chicago_crime.crime`
    WHERE location_description = 'ALLEY'
    """
rt6=client.query(query6).to_dataframe()
rt6
```

### GROUP BY 
GROUP BY takes the name of one or more columns, and treats all rows with the same value in that column as a single group when you apply aggregate functions like COUNT().
We doesn't use GROUP BY without an aggregate function. Similarly, if you have any GROUP BY clause, then all variables must be passed to either a GROUP BY command, or an aggregation function.

```python
query7="""
    SELECT COUNT(case_number)
    FROM `bigquery-public-data.chicago_crime.crime`
    GROUP BY arrest
    """
rt7=client.query(query7).to_dataframe()
rt7
```

### HAVING
HAVING is used in combination with GROUP BY to ignore groups that don't meet certain criteria

```python
query8="""
    SELECT COUNT(case_number)
    FROM `bigquery-public-data.chicago_crime.crime`
    GROUP BY arrest
    HAVING COUNT(unique_key)>1000
    """
rt8=client.query(query8).to_dataframe()
rt8
```

### ORDER BY

```python
query9="""
    SELECT unique_key,case_number
    FROM `bigquery-public-data.chicago_crime.crime`
    ORDER BY unique_key
    """
rt9=client.query(query9).to_dataframe()
rt9
```

```python
query10="""
    SELECT unique_key,case_number
    FROM `bigquery-public-data.chicago_crime.crime`
    ORDER BY unique_key DESC
    """
rt10=client.query(query10).to_dataframe()
rt10
```

### Dates

```python
# we can extract DAY,WEEK(week in the year between 1 and 53),DAYOFWEEK(1 TO 7 (1 (Sunday) and 7 (Saturday))) from DATE[YY-MM-DD] or DATETIME
# extract day from date
query11="""
    SELECT unique_key,EXTRACT(DAY from date)
    FROM `bigquery-public-data.chicago_crime.crime`
    """
rt11=client.query(query11).to_dataframe()
rt11
```

```python
# extract week from date
query12="""
    SELECT unique_key,EXTRACT(DAYOFWEEK FROM date)
    FROM `bigquery-public-data.chicago_crime.crime`
    WHERE year=2010
    """
rt12=client.query(query12).to_dataframe()
rt12
```

### WITH AS - Common Table Expression (CET)
A CTE is a temporary table that you return within your query. CTEs are helpful for splitting your queries into readable chunks, and you can write queries against them.


```python
query00="""
    WITH casedetails AS
    (
    SELECT unique_key,case_number
    FROM `bigquery-public-data.chicago_crime.crime`
    WHERE fbi_code= '01A'
    )
    SELECT case_number
    FROM casedetails
    """
rt00=client.query(query00).to_dataframe()
rt00
```
